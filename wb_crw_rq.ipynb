```
#응답 코드를 활용한 조건 분기
import requests as rq

def url_check(url):
  res = rq.post(url)
  print(res)

  sc=res.status_code

  if sc==200:
    print("%s request sucess" %(url))

  elif sc ==400:
    print("%s not found" %(url))

  else:
    print("%s unkowned error: %s" %(url, sc))

url_check("https://github.com/minibe0")
url_check("https://github.com/minibe0//aa")
```

```
#데이터 보내기
#쿼리스트링 데이터 만들어 요청
#params를 사용

import requests as rq

url = "https://minibe0.github.com"

res= rq.get(url, params={"key1":"value1", "key2":"value2"})

print(res.url)
```

```
#post 요청
#post를 요청할 때는 데이터가 url에 포함되어 있지 않고 
#header의 body에 포함되기 때문에
#반드시 추가적인 인자를 넣어 보내야 한다.

import requests as rq

url = "http://www.example.com"

res=rq.post(url, data={"key1":"value1", "key2":"value2"})

print(res.url)
```

```
#json 모듈을 활용한 post 요청
#json 을 import한 후 json.dump(딕셔너리)를 하면 
#해당 딕셔너리의 형태를 유지하면서 문자열로 바꾸어 준다


import requests as rq
import json
url = "http://www.example.com"

res = rq.post(url, data=json.dumps({"key1":"value1", "key2":"value2"}))

print(res.url)

#결과가 달라지지는 않았지만 data를 딕녀서리에서 문자열 형태로 바꾸어 보낸 것이다
```
